# Look on your composite styles

## choose your style out of a list of available styles

```python
# show styles the ./styles folder

import os
import shutil

# get file names of all styles
styles = os.listdir('styles')
styles.sort()

# order matters
styles_to_use = [3, 1]

for i in range(len(styles)):
    if i in styles_to_use:
        print(i, "<"+13*'-', styles[i])
    else:
        print(i, "\t\t", styles[i])
```

```python
style_names = [os.path.join('styles', styles[i]) for i in styles_to_use]


import matplotlib.pyplot as plt
import numpy as np

# set style
plt.style.use(style_names)
```

# plot example

```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(9*x) * np.cos(0.5*x)
x2 = np.linspace(0, 10, 30)
y2 = np.exp(-0.4*x2) * np.sin(2*x2)
y3 = 0.08 * x + 0.1

y2_std = np.abs(np.random.normal(0, 0.1, len(y2)))
y3_std = np.abs(np.random.normal(0.043, 0.006, len(y3)))

plt.title('Example plot')
plt.plot(x, y1, label='signal')
plt.errorbar(x2, y2, yerr=y2_std, fmt='o', label='signal with errors')
plt.plot(x, y3, label='fit')
plt.fill_between(x, y3 - y3_std, y3 + y3_std, alpha=0.3, label='fit error', color='C2')
plt.xlabel('time (s)')
# plt.ylabel('y')
plt.legend()
plt.show()

```

## Histogram example

```python
# generate random data
x = np.random.normal(0, 1, 1000)
y = np.random.normal(2, 0.5, 1000)

# plot histogram
plt.hist(x, bins=30, label='first')
plt.hist(y, bins=30, label='second')
plt.xlabel('x')
plt.ylabel('count')
plt.legend()
plt.show()
```

## image example

```python
# generate 2d gaussian data
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.5*((X-0.7)**2 + (Y-0.3)**2))

# plot 2d gaussian data
plt.figure(figsize=(2.6, 2.6))
plt.pcolormesh(X, Y, Z, vmax=1, vmin=0)
plt.gca().set_aspect(1.0)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# plot with imshow

# plot 2d gaussian data
plt.imshow(Z, vmax=1, vmin=0, origin='lower', extent=[-3, 3, -3, 3])
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
